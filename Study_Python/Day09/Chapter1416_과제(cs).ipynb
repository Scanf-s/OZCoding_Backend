[문제1]

파이썬 2.7 버전과 파이썬 3에서 range 차이를 설명해주세요

Python 2.7의 range는 고정된 메모리를 할당하여 list를 생성하여, 크기가 클 수록 많은 양의 메모리가 필요합니다.
또 Python 2.7에는 xrange라는 기능이 존재했는데, 이는 range()의 단점인 고정 할당 메모리 방식에서 벗어나, 동적 메모리 할당 방식을 사용하는 함수입니다. 즉, 필요한 만큼 메모리를 사용하여 리스트를 생성할 수 있습니다.

Python 3의 range는 Python 2.7의 **xrange**와 비슷하게 기능합니다.
즉, 2.7의 xrange를 Python 3에서 range로 사용할 수 있게 되었다는 것입니다. 따라서, 동적 메모리 할당 리스트를 생성해줍니다.

따라서 Python 2.7의 range는 메모리 사용시, 효율성이 좋지 못하지만  
Python 3의 range는 효율적으로 메모리를 사용합니다.

[문제2]

Python의 얕은 복사와 깊은 복사에 대해서 설명하세요

컴퓨터 언어에서는
어떠한 객체 또는 자료구조가 있을 때, 이를 가리키는 주소값이 있습니다.
모든 변수, 객체, 자료구조(이하 Object)는 이 주소값을 통해 접근하여 값을 읽고 쓸 수 있습니다.

따라서 얕은 복사는 Object의 복사본을 생성하지만 복사본과 원본 객체가 요소들을 공유하게 됩니다.  
즉, Object의 복사본을 생성할 때, 객체의 최상위 수준만 (Object 자체의 주소값) 새로 생성되고, 내부에 있는 하위 객체들의 참조(주소값)는 그대로 유지됩니다.  
따라서, 복사본이나 원본 중 하나를 수정하면, 내부 객체가 공유되기 때문에 서로에게 영향을 줄 수 있습니다.

깊은 복사는 Object의 복사본을 생성할 때, 원본 Object와 분리된 복사본(새로운 주소값 할당)을 생성합니다.  
이 과정에서 원본 Object뿐만 아니라 내부에 포함된 모든 elements도 새롭게 복사되어, 복사본의 구조 전체가 원본과 만들어집니다.
따라서, 깊은 복사를 사용하면 원본 Object를 변경하더라도 복사본에는 영향을 주지 않고, 복사본을 변경해도 원본 Object에는 영향을 주지 않습니다.
