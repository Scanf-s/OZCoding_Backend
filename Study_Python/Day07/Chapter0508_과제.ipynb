[문제1]

변수명 게임에 들어 있는 값 중 '홀'만 출력되도록 코드를 작성해 주세요

게임 = "홀짝홀짝홀짝"

출력 예시 : "홀홀홀"


```python
게임 = "홀짝홀짝홀짝"
result = ""
for element in 게임:
    if element == "홀":
        result += element
print(result)
```

[문제2]

변수명 전화번호에 들어 있는 번호를 모두 붙여서 출력되도록 코드를 작성해주세요

전화번호 = "010-1234-5678"

출력 예시 : "01012345678"

```python
전화번호 = "010-1234-5678"
result = ""

for 문자 in 전화번호:
    if 문자.isdigit():  # 문자가 숫자인지 확인
        result += 문자

print(result)
```

[문제3]

창의성, 시장성, 안정성, 수익성에 대한 각각의 평가 중 아래 기준에 부합하지 못하면 창업지원을 받을 수 없습니다.

- 수익성 60점 이상
- 창의성과 시장성 50점 이상
- 시장성과 수익성을 더한 값의 절반이 안정성 보다 높아야 함

다음 점수가 입력되었을때 창업 지원이 가능한지 여부를 True 또는 False로 출력해주세요

---

창의성 : 62
시장성 : 71
안정성 : 68
수익성 : 66

```python
score = {"창의성": 62, "시장성": 71, "안정성": 68, "수익성": 66}
def qualifyCheck(score):
    if(score["수익성"] >= 60 and score["창의성"] >= 50 and score["시장성"] >= 50 and (score["시장성"] + score["수익성"] / 2) > score["안정성"]):
        return True
    
    
if(qualifyCheck):
    print(True)
else:
    print(False)
```
