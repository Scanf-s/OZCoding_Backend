```python
# Classification
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
```

```python
iris.data
```

```python
iris.target
```

```python
# 준비된 데이터셋을 분류한다 학습용 : (train_data) // 테스트용 (test_data)

from sklearn.model_selection import train_test_split
X_trian, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=123)
```

```python
y_test
len(y_train)
```

```python
# 학습을 위한 알고리즘 불러오기
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(random_state=123)
dt_clf
```

```python
# 학습 진행

dt_clf.fit(X_trian, y_train)
```

```python
# 테스트

predict = dt_clf.predict(X_test)
predict
```

```python
# 결과 확인

from sklearn.metrics import accuracy_score

accuracy_score(y_test, predict)
```

```python
df = pd.DataFrame(data=X_test, columns=iris.feature_names)

df["answer"] = y_test
df["predict"] = predict

df
```
