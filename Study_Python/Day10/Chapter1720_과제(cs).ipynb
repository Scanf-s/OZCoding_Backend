객체지향에 대해서 설명해주세요

객체지향 프로그래밍이란, 실세계의 객체를 프로그래밍 언어로 표현하기 위한 하나의 수단이다.
객체지향을 통해 객체의 속성과 하는 행동을 지정할 수 있으며, 객체의 하위 집합(상속)에 포함되는 객체까지 수월하게 표현 가능하다

1) 추상화
- 어떤 객체의 본질적인 특징을 정의할 수 있다
2) 상속
- 기존 객체를 활용하여 하위 객체를 정의할 수 있다.
3) 다형성
- 상속 및 인터페이스 구현 시 메서드 오버로딩, 오버라이딩을 통해 여러 기능 및 형태로 변형 가능하다
4) 캡슐화
- 클래스의 속성 및 메서드를 묶어서 외부로부터의 접근을 최소화하는 것 (private, protected, default, public)

[문제4]

클래스 다중 상속과 관련된 MRO에 대해 설명해주세요

MRO : 클래스에서 메서드 호출 시 Python이 해당 메서드를 찾기 위해 따르는 Protocol

클래스 다중 상속이란 한 클래스가 두개 이상의 상위 클래스로부터 상속을 받는 것인데, 이러한 상황에서
만약 두 상위 클래스에 중복된 이름의 함수가 존재하는 경우 어떤 부모 클래스의 메서드를 사용해야 하는지 결정해야 한다.

Python2에서는 MRO 처리 시, DFS를 사용하였으나, 이후 복잡한 문제 상황을 개선하기 위해 점차 **C3 선형화 알고리즘**을 도입하게 되었다.
Python3에서는 모든 클래스에서 **C3 선형화 알고리즘**을 사용하여 MRO를 계산한다.

**C3 선형화 알고리즘**은, 다음 세 규칙을 반드시 따른다.

1. 하위 클래스는 상위 클래스보다 먼저 나타난다.
- 즉, 하위 클래스가 상위 클래스의 메서드를 오버라이드 할 수 있으니까 하위 클래스에 오버라이딩 된 메서드가 우선적으로 고려된다. 이는 Java와 마찬가지인데, Java에서도 하위 클래스에 Override된 메서드 우선으로 호출한다.

2. 부모 클래스의 순서는 유지된다.
- 하위 클래스를 정의할 때 상위 클래스를 상속하기 위해 매개변수로 받게 되는데, 이 때 나열한 순서가 메서드 중복 해결 순서에도 반영된다.

3. 유효한 선형 순서는 하나만 존재해야 한다.
- Class 계층에는 모순이 존재해서는 안되며, 모든 클래스는 선형 순서에서 단 한번만 나타나야만 한다.
즉, 상위부터 하위까지 쭉 타고 내려올 때, 중복되는 이름이 존재하면 안된다는 의미이다.

Python에서 MRO를 확인하려면,

```python
클래스명.mro()
help(클래스명)
```
위 두가지 함수를 통해 MRO를 확인할 수 있다.

[문제6]

ZeroDivisionError,  IndexError 외에 3가지 이상의 파이썬 에러에 대해 설명해주세요

모든 Exception은 BaseException 클래스로부터 상속받아 정의되었으며,
예를 들어 다음과 같은 에러가 존재한다.

1. KeyError : Hashmap(Dictionary)사용 시 키가 존재하지 않는 경우
2. ValueError : 연산이나 함수가 올바른 타입이지만, 거기에 저장된 값이 타입에 맞지 않는 부적절한 값인 경우
3. NameError : 로컬 또는 글로벌 변수에서 해당하는 이름을 가진 변수를 찾을 수 없는 경우
4. TypeError : 연산이나 함수가 부적절한 타입의 객체에 적용되었을 때 발생
5. FileNotFoundError : 요청된 파일 또는 디렉토리가 존재하지 않는 경우
