[문제2]

Python의 실수형(float) 계산 시 오차가 발생하는 이유에 대해 간단히 설명해 주세요

컴퓨터는 정수를 저장할때와는 달리, 실수를 실제 값이 아닌, 근사값으로 저장합니다.
왜 근사값으로 저장할까요? 실수는 부호비트, 지수부, 가수부로 저장이 되는데, 지수부는 수의 정밀도를 결정하는 부분이고, 가수부는 실수의 표현 범위를 지정하는 부분입니다. 이때, 1/3과 같은 순환소수는 지수부와 가수부의 크기가 제한되어 있기 때문에 절대로 실제 값을 저장할수가 없습니다.

따라서, 1.1 + 0.1 == 1.2가 False인 것 처럼, 실수 계산시에는 실제 우리가 눈으로 보는 값이 아닌, 근사값끼리의 연산이기 때문에 당연히 오차가 발생합니다.
